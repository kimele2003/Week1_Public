### Univariate Analysis:

```python
"""Let's revisit the Dataset_1 again. Read the csv file https://raw.githubusercontent.com/bwsimedlytics/Week1Public/master/BWSI_Dataset1.csv into a dataframe
and print the first 7 rows"""

## YOUR CODE HERE
url = 'https://raw.githubusercontent.com/bwsimedlytics/Week1Public/master/BWSI_Dataset1.csv'
# names = ['Year','Sex','Age','Ethnicity','ParentEducation','MaritalStatus','HouseIncome','Protein','Carb','TotalFat','Chol','VitA','VitC','Weight','Height','BMI','BloodPres','Diabetes','HealthIns','HoursTV']
data = pd.read_csv(url)
data.head(7)
```

```python
"""Conduct a Univariate Analysis of the participants' Cholesterol. What is noticeable about it? Create a function
that makes the participants' Cholesterol more Normal and create two visualizations, 
one before the Chol data is modified, and one after Chol data is nmodified to be more Normal than before."""

## YOUR CODE HERE

# data['Chol'].describe()
# sns.boxplot('Chol', data=data)

fig, ((axs1, axs2), (axs3, axs4)) = plt.subplots(2,2)
axs1.boxplot(data['Chol'])
plt.figure(1)
axs2.hist(data['Chol'])

def more_Normal(dataframe, column): 
  ## YOUR CODE HERE
  # from Matplotlib_Soutions.ipynb:
  y = dataframe[column]
  y = y[y.between(y.quantile(.15), y.quantile(.85))]
  return y

axs3.boxplot(more_Normal(data,"Chol"))
axs4.hist(more_Normal(data,"Chol")) #better representation of data, more Normal than before, now not as strongly skewed

# I wrote 'axs4' instead of 'plot' in the line above

#Hint: which data visualization shows the outliers?
```

```python
"""Conduct a Bivariate Analysis for Protein and Chol. What is familar about this compared to the tutorial? What is another 
way to deal with overplotting other than the Hex plot? Create two data visualizations, one that shows the overplotting, the other that 
resolves this issue. After you are done, play around with the sns.kdeplot function or try to create a 3D graph to reduce the effects of overplotting."""

## YOUR CODE HERE
x = data['Protein']
y = data['Chol']
plt.figure(1)
plt.scatter(x,y)
plt.xlabel('Protein')
plt.ylabel('Chol')
plt.title('Overplotted Scatter Plot of Protein vs. Chol') # shows overplotting

rand_samp = data.sample(500)
x = rand_samp['Protein']
y = rand_samp['Chol']
plt.figure(2)
plt.scatter(x,y,color='red')
plt.xlabel('Protein')
plt.ylabel('Chol')
plt.title('Random Sampled Scatter Plot of Protein vs. Chol') # fixes overplotting
```
